```python
import pandas as pd
```

```python
ratings = pd.read_excel("/content/drive/MyDrive/Spechio_Project/SCUT-FBP5500_v2/All_Ratings.xlsx")
```

```python
ratings_avg = ratings.groupby('Filename')['Rating'].mean().reset_index()
ratings_avg.shape
```

```python
ratings_avg.head(5)
```

```python
gr = pd.read_csv("/content/drive/MyDrive/Spechio_Project/golden_ratios.csv")
```

```python
sr = pd.read_csv("/content/drive/MyDrive/Spechio_Project/symmetric_ratios.csv")
```

```python
glcm = pd.read_csv("/content/drive/MyDrive/Spechio_Project/glcm_features.csv")
```

```python
hu = pd.read_csv("/content/drive/MyDrive/Spechio_Project/hu_moments.csv")
```

```python
gr.head(5)
```

```python
sr.head(5)
```

```python
glcm.head(5)
```

```python
hu.head(5)
```

```python
features = pd.merge(gr,sr,on='Image Name',how='left')
features
```

```python
features = pd.merge(features,glcm,on='Image Name',how='left')
features
```

```python
features = pd.merge(features,hu,on='Image Name',how='left')
features
```

```python
#Final Dataframe Containing 27 Features and Ratings
df = pd.merge(features, ratings_avg, left_on='Image Name', right_on='Filename', how='left').drop('Filename', axis=1)
df
```

```python
# Save the dataframe as a CSV file
df.to_csv('/content/drive/MyDrive/extracted_features.csv', index=False)
```
